# Learning Algorithm

The agent is using basic DQN algorithm to solve the navigation task. It uses a replay buffer to store past experience and uses batches of experiences to train the model

List of hyperparameters:
- BUFFER SIZE = int(1e5)  (replay buffer size)
- BATCH_SIZE = 64         (minibatch size)
- GAMMA = 0.99            (discount factor)
- TAU = 1e-3              (for soft update of target parameters)
- LR = 5e-4               (learning rate )
- UPDATE_EVERY = 4        (how often to update the network)

In DQN, the Q table is estimated by using a neural net (NN). In this perticular implementation the NN is made of 3 fully connected layers:
The 1st layer has 37 inputs (the state space of the environment) and 64 outputs
The 2nd layer has 64 inputs and 64 outputs
The final layer has 64 inputs and 4 outputs (One for the value of each actions)
